```python
import pandas as pd
```

```python
df = pd.read_csv('311-truncated.csv')
```

```python
df.head()
```

```python
df['Complaint Type'].value_counts()
```

```python
complaints = df['Complaint Type'].value_counts()
x = complaints.index
y = complaints
py.iplot([Bar(x=list(x), y=y)], filename='complaint type')
```

```python
agency = df['Agency'].value_counts()
x = agency.index
y = agency
py.iplot([Bar(x=list(x), y=y)], filename='agency type')
```

```python
df['Agency'].str.contains('NYPD')
```

```python
idx = df['Agency'].str.contains('NYPD')
```

```python
complaints = df['Complaint Type'][idx].value_counts()
x = complaints.index
y = complaints
py.iplot([Bar(x=list(x), y=y)], filename='NYPD complaints')
```

```python
df['Agency'].unique()
```

```python
nypd = df[df['Agency'].str.contains('NYPD')]['Complaint Type'].value_counts()[0:30]
fdny = df[df['Agency'].str.contains('FDNY')]['Complaint Type'].value_counts()[0:30]
```

```python
nypd_bar = Bar(x=list(nypd.index), y=nypd, name='NYPD')
fdny_bar = Bar(x=list(fdny.index), y=fdny, name='FDNY', xaxis='x2', yaxis='y2')

layout = Layout(
    yaxis=YAxis(
        domain=[0, 0.45]
    ),
    yaxis2=YAxis(
        domain=[0.55, 1]
    ),
    xaxis2=XAxis(
        anchor='y2'
    ),
    legend=Legend(
        traceorder='reversed'
    )
)

fig = Figure(data=[nypd_bar, fdny_bar], layout=layout)

py.iplot(fig)
```

```python
unique_agencies = df['Agency'].unique()
fig = tls.get_subplots(rows=len(unique_agencies), vertical_spacing=1./(3*len(unique_agencies)))
```

```python
print fig.to_string()
```

```python
for i, agency in enumerate(unique_agencies):
    complaints = df[df['Agency'].str.contains(agency)]['Complaint Type'].value_counts()[0:30]
    bar = Bar(x=list(complaints.index), y=list(complaints), name=agency, xaxis='x'+str(i+1), yaxis='y'+str(i+1))
    fig['data'].append(bar)
```

```python
print fig.to_string()
```

```python
py.iplot(fig, height=5000, filename='NYC 311 complaints by agency')
```
